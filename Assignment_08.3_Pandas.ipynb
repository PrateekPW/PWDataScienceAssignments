# Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

loc is used to access a group of rows and columns by labels or a boolean array.
iloc is used to access a group of rows and columns by integer position.

# Q3. Reindex the given dataframe using a variable, reindex = [3,0,1,2] and store it in the variable, new_df then find the output for both new_df.loc[2] and new_df.iloc[2].

```python
import pandas as pd

# Creating the DataFrame
course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})
print("Original DataFrame:\n", df)

# Reindexing the DataFrame
reindex = [3, 0, 1, 2]
new_df = df.reindex(reindex)
print("\nReindexed DataFrame:\n", new_df)

# Output of new_df.loc[2]
loc_output = new_df.loc[2]
print("\nOutput of new_df.loc[2]:\n", loc_output)

# Output of new_df.iloc[2]
iloc_output = new_df.iloc[2]
print("\nOutput of new_df.iloc[2]:\n", iloc_output)
```

new_df.loc[2] accesses the row with the label 2, which corresponds to the original index 2 of the DataFrame. After reindexing, this row contains the values course_name: Big Data and duration: 6.

new_df.iloc[2] accesses the row at the integer position 2 in the new DataFrame. This corresponds to the third row of the reindexed DataFrame, which originally had the index 1, containing the values course_name: Machine Learning and duration: 3.

## Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days, hours, and minutes using Pandas time delta. The program should prompt the user to enter the dates and display the result.

```python
def calculate_time_difference(date1_str, date2_str):
    # Convert the input strings to Timestamps
    date1 = pd.Timestamp(date1_str)
    date2 = pd.Timestamp(date2_str)
    
    # Calculate the difference
    time_difference = date2 - date1
    
    # Extract the days, hours, and minutes
    days = time_difference.days
    hours, remainder = divmod(time_difference.seconds, 3600)
    minutes, _ = divmod(remainder, 60)
    
    return days, hours, minutes

# Prompt the user to enter the dates
date1_str = input("Enter the first date (YYYY-MM-DD): ")
date2_str = input("Enter the second date (YYYY-MM-DD): ")

# Calculate the time difference
days, hours, minutes = calculate_time_difference(date1_str, date2_str)
```

## Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified column to a categorical data type. The program should prompt the user to enter the file path, column name, and category order, and then display the sorted data.

```python
df = pd.read_csv('weather.csv')
df
```

```python
import pandas as pd
def read_and_convert_to_categorical(file_path, column_name, category_order):
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    # Convert the specified column to a categorical data type with the given order
    df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)
    
    # Sort the DataFrame by the specified categorical column
    df_sorted = df.sort_values(by=column_name)
    
    return df_sorted

# Prompt the user to enter the file path, column name, and category order
file_path = input("Enter the file path of the CSV file: ")
column_name = input("Enter the column name to convert to categorical: ")
category_order = input("Enter the category order (comma-separated values): ").split(',')

# Read, convert, and sort the data
df_sorted = read_and_convert_to_categorical(file_path, column_name, category_order)

# Display the sorted data
print("\nSorted DataFrame:\n", df_sorted)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('sales_data.csv')
    
    # Ensure the 'Date' column is of datetime type
df['Date'] = pd.to_datetime(df['Date'])
    
    # Pivot the data to get product categories as columns and dates as rows
sales_pivot = df.pivot_table(index='Date', columns='Product', values='Order_Quantity', aggfunc='sum')
    
    # Plot the data
sales_pivot.plot(kind='bar', stacked=True, figsize=(10, 7))
    
    # Set the title and labels
plt.title('Sales of Each Product Category Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
    
    # Show the legend and plot
plt.legend(title='Product')
plt.show()
```
