Q2: What is a scatter plot?

Ans A scatter plot is a type of data visualization that displays values for typically two variables for a set of data. It uses Cartesian coordinates to display points, where each point represents an observation in the dataset. The position of each point on the horizontal and vertical axis indicates values for an individual data point.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

 Use the following data:
 
 import numpy as np
 
 For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])
 
 For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
 
 For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
 
 For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
 
 
 Ans The subplot() function in Matplotlib is used to create multiple plots in a single figure. This allows for better comparison of different datasets and more efficient use of space in the graphical representation.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure
plt.figure(figsize=(10, 8))

# First subplot
plt.subplot(2, 2, 1)
plt.plot(x1, y1, 'r') # 'r' is the color red
plt.title('Line 1')
plt.xlabel('x')
plt.ylabel('y')

# Second subplot
plt.subplot(2, 2, 2)
plt.plot(x2, y2, 'g') # 'g' is the color green
plt.title('Line 2')
plt.xlabel('x')
plt.ylabel('y')

# Third subplot
plt.subplot(2, 2, 3)
plt.plot(x3, y3, 'b') # 'b' is the color blue
plt.title('Line 3')
plt.xlabel('x')
plt.ylabel('y')

# Fourth subplot
plt.subplot(2, 2, 4)
plt.plot(x4, y4, 'm') # 'm' is the color magenta
plt.title('Line 4')
plt.xlabel('x')
plt.ylabel('y')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

 Q4: What is a bar plot? Why is it used?
 
 A bar plot, also known as a bar chart, is a type of graph that represents categorical data with rectangular bars. Each bar's length or height is proportional to the values it represents. Bar plots can be plotted either vertically or horizontally.

Uses of a Bar Plot:

Comparison: Bar plots are excellent for comparing different categories or groups. For example, comparing sales figures across different regions or time periods.

Trend Analysis: They can be used to identify trends over time when categories are ordered in a meaningful sequence, such as months or years.

Distribution: Bar plots can display the frequency distribution of a categorical dataset, showing how many occurrences there are for each category.

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.subplot(2, 2, 1)
plt.bar(company,profit)
plt.subplot(2, 2, 3)
plt.barh(company,profit)
```

 Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
 
 A box plot, also known as a box-and-whisker plot, is a standardized way of displaying the distribution of data based on a five-number summary: 
 minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It helps in visualizing the spread and skewness of the data set.

 

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2],patch_artist=True)
```
